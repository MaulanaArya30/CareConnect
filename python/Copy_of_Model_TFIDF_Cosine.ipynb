```python
def clean_list(jenis):
  texts = jenis.split(",")
  lentxt = len(texts)
  if lentxt <= 0:
    return []
  for i in range(lentxt):
    texts[i] = texts[i].replace("'","").replace("[","").replace("]","").strip()
  return texts

df['Jenis Yayasan'] = df.apply(lambda x: clean_list(x['Jenis Yayasan']), axis=1)
df['Tag Kebutuhan'] = df.apply(lambda x: clean_list(x['Tag Kebutuhan']), axis=1)
df['Tag Program'] = df.apply(lambda x: clean_list(x['Tag Program']), axis=1)
df
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import random
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopworda = set(stopwords.words('indonesian'))
```

```python
def clean_text(text):
    temp = text.lower()
    temp = re.sub("'", "", temp) # to avoid removing contractions in english
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    temp = re.sub("[^a-z0-9]"," ", temp)
    temp = temp.split()
    temp = [w for w in temp if not w in stopworda]
    temp = " ".join(word for word in temp)
    return temp
```

```python
df['Description_clean'] = df.apply(lambda x: clean_text(x['Description']), axis=1)
df['Kebutuhan_clean'] = df.apply(lambda x: clean_text(x['Kebutuhan']), axis=1)
df['Program_clean'] = df.apply(lambda x: clean_text(x['Program']), axis=1)
df
```

```python
df['text_temp'] = df['Nama Panti'] + " " + df['Description_clean'] + " " + df['Kebutuhan_clean'] + " " + df['Program_clean']
df
```

```python
def combine_text_need(text, needs):
  for need in needs:
    text = text + f" {need}"
  return text

df['text'] = df.apply(lambda x: combine_text_need(x['text_temp'], x['Tag Kebutuhan']), axis=1)
df.drop(['text_temp'], axis = 1, inplace=True)
```
